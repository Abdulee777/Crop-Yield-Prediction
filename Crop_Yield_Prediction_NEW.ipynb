# DATA ENCODING

```python
# Define a mapping from 'rice' to 'zero' and 'maize' to '1'
crop_mapping = {'Rice': 0, 'Maize': 1}
# Use the map function to apply the mapping to the 'Crop' column
data['Crop'] = data['Crop'].map(crop_mapping)


# Define a mapping soil type where "clay"= 0, "loamy"= 1 and  "sandy"= 2
soil_mapping = {'clay': 0, 'loamy': 1, 'sandy':2}
# Use the map function to apply the mapping to the 'Crop' column
data["Soil Type"] = data["Soil Type"].map(soil_mapping)


# Define a mapping pesticides where "gammalin-20"= 0, "Nogos-50"= 1 and  "Vetox-85"= 2
pesticides_mapping = {'Gammalin 20': 0, 'Nogos 50': 1, 'Vetox 85':2}
# Use the map function to apply the mapping to the 'Crop' column
data["Pesticides"] = data["Pesticides"].map(pesticides_mapping)


herbicides_mapping = {'PARAFORCE (PARAQUAT)':1, 'FORCE UP (GLYPHOSATE)':2, 'RiceForce (oxadiazon)':3, '2,4-D':4,
           'ATRAFORCE (ATRAZINE)':5,  'BUTACHLOR':6, '3MAIZEFORCE':7, 'Propalnil + 2.4 D':8, 'Bispyribac Sodium':9,
           'POWERFORCE (Atrazine + Paraquat)':10, 'BentaForce':11,
           'GuardForce':12, 'ATALAR':13, 'METOLACHLOR':14
           }
# Use the map function to apply the mapping to the 'Crop' column
data["Herbicides"] = data["Herbicides"].map(herbicides_mapping)


seed_mapping = {'Faro-55':1, 'Faro-54':2, 'Faro-44':3, 'Gawal-R1':4, 'Faro-24':5, 'Faro-62':6,
           'SC-419':7,'DK-649':8, 'DK-234':9, 'DK-777':10, 'SC-612':11, 'Oba-98':12}
# Use the map function to apply the mapping to the 'Crop' column
data["seed_type"] = data["seed_type"].map(seed_mapping)

season_mapping = {'Rainy':0, 'Dry':1}
# Use the map function to apply the mapping to the 'Crop' column
data["Season"] = data["Season"].map(season_mapping)


fert_mapping = {'NPK 15:15:15':0, 'NPK 12:12:17':1, 'NPK 20:10:10':2, 'NPK 2-1-1':3,
       'NPK 12-11-12':4, 'NPK 27:13:13':5}
# Use the map function to apply the mapping to the 'Crop' column
data["Fertilizer"] = data["Fertilizer"].map(fert_mapping)
```

```python
X= data.drop(["Year", "Crop Yield (kg/ha)"], axis=1)

y= data['Crop Yield (kg/ha)']
```

```python

import sklearn

sklearn.__version__
```

```python
random_seed=50

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test= train_test_split(X, y, test_size=0.2, random_state=random_seed)
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, train_test_split
import matplotlib.pyplot as plt

from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, mean_absolute_percentage_error
from sklearn.metrics import r2_score

import time

import warnings
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
from sklearn.feature_selection import RFE
```

```python
model = LinearRegression()
start_time_train = time.time()

rfe = RFE(estimator=model, n_features_to_select=5)
rfe = rfe.fit(x_train, y_train)

end_time_train = time.time()
linear_training_time = end_time_train - start_time_train

# Get the ranking of features
ranking = rfe.ranking_
selected_features = X.columns[rfe.support_]

print("Selected Features:", selected_features)
print("Ranking:", ranking)

# Step 6: Use the Selected Features
X_selected = X[selected_features]

# Optional: Train a model with the selected features
model.fit(X_selected, y)

start_time_test = time.time()
pred= model.predict(x_test[selected_features])
end_time_test = time.time()
linear_testing_time = end_time_test - start_time_test
mae_linear= mean_absolute_error(y_test, pred)
mse_linear= mean_squared_error(y_test,pred)
rmse_linear= mean_squared_error(y_test, pred, squared=False)
rmse_linear=np.sqrt(rmse_linear)
r2_linear= r2_score(y_test, pred)

# Calculate the range of the target variable
target_range = np.max(y_train) - np.min(y_train)

# Calculate RMSE as a percentage of the target range
rmse_percentage = (rmse_linear / target_range) * 100

print("RMSE percentage: {:.2f}%".format(rmse_percentage))

r2_percentage = r2_linear * 100
print("R-squared (R2) Score: {:.2f}%".format(r2_percentage))

print(f"MAE Linear: {mae_linear},MSE Linear:{mae_linear},RMSE Linear: {rmse_linear},R2 Score Linear: {r2_linear}")
print(f"Training time: {linear_training_time} seconds")
print(f"Testing time: {linear_testing_time} seconds")
```

```python
rand = RandomForestRegressor(n_estimators=100, random_state=random_seed)
rfe = RFE(model, n_features_to_select=2)
rfe.fit(x_train, y_train)
print("Feature Ranking: ", rfe.ranking_)
print("Selected Features: ", rfe.support_)
x_train_rfe = rfe.transform(x_train)
x_test_rfe = rfe.transform(x_test)
model.fit(x_train_rfe, y_train)
score = model.score(x_test_rfe, y_test)
print("Model accuracy with selected features: ", score)


```

```python
# Create a Linear Regressor
reg = LinearRegression()

# Perform cross-validation
scores = cross_val_score(reg, X, y, cv=5)
cv_scores_percentage = scores * 100
cv_scores_percentage_scalar = np.mean(cv_scores_percentage)  # Convert to scalar (e.g., mean)
print("Cross-Validation Scores: {:.2f}%".format(cv_scores_percentage_scalar))

# Print cross-validation scores
print("Cross-validation scores:", scores)
print("Mean cross-validation score:", np.mean(scores))
print("Mean R2 Score:", scores.mean())
print("Standard Deviation of R2 Scores:", scores.std())
```
